## <span style="color: teal;">Working with Tuples</span>
### <span style="color: pink;">Characteristics of Tuples</span> 
- They are ordered.
- They are NOT mutable aka immutable.
- They can contain multiple data types.
- They are indexable and slicable like lists.
- They support nesting.

```python
# Create a tuple of five integers and assign it to the variable tuple_1
print("1. Create a tuple of five integers and assign it to the variable tuple_1")
tuple_1 = (1, 2, 3, 4, 5)
tuple_1 
print(" ")

# Print the type of the tuple
print("2. Type of the tuple")
print(type(tuple_1))
print(" ")

# Create a tuple with the elements 1, 2, "three", and the list 'Python' and assign it to the variable tuple_2
print("3. Create a tuple with the elements 1, 2, 'three', and the list 'Python'")
tuple_2 = (1, 2, "three", list('Python'))
tuple_2
print(" ")

# Retrieve the first element of tuple_1
print("4. Retrieve the first element of tuple_1")
print(tuple_1[0])
print(" ")

# Create a typle with the strign elements 'Python', 'R', and 'Julia' and assign it to languages
print("5. Create a tuple with the string elements 'Python', 'R', and 'Julia'")
languages = ('Python', 'R', 'Julia')
languages
print(" ")

# Assign each element of the tuple to a separate variable using the variable names one, two, three
print("6. Assign each element of the tuple to a separate variable using the variable names one, two, three")
one, two, three = languages
print(" ")

# Print the variable one
print("7. Print the variable one")
print(one)
print(" ")

#Creating lists
print("8. Creating lists")
names = ['Mary', 'Anne', 'Rene']
surnames = ['Johnstone', 'Barnes', 'Le Roux']
print(" ")

# Use the zip function to create a list of tuples with the names and surnames
print("9. Use the zip function to create a list of tuples with the names and surnames")
names_surnames = list(zip(names, surnames))
print(names_surnames) 
print(" ")

# Generate a list of tuples (each with three elements)
print("10. Generate a list of tuples (each with three elements)")
data_set = [('Anna', 'Rogers', 33), ('Susan', 'Roberts', 25), ('Rene', 'Du Bois', 28)]
print(data_set)
print(" ")

# Unzip the three element tuples
print("11. Unzip the three element tuples")
names, lastnames, ages = zip(*data_set)
print(" ")

# Return the names
print("12.mReturning the names")
print(names)
print(" ")

# Return the lastnames
print("13. Returning the lastnames")
print(lastnames)
print(" ")

# Return the ages
print("14. Returning the ages")
print(ages)
print(" ")

# Create a tuple named primes containing the first seven prime numbers
print("15. Create a tuple named primes containing the first seven prime numbers")
primes = (2, 3, 5, 7, 11, 13, 17)
print(" ")
# Using indexing return the last three primes in the tuple
print(primes[-3:])
print(" ")


```

# <span style="color: teal;">Working with Dictionaries</span>
### <span style="color: pink;">Characteristics of Dictionaries</span> 
- They have key-value pairs.
- They are mutable. Keys and values can be added, modified, or deleted.
- Keys must be unique and of immutable types (string, number, tuple).
- Iterable ove keys, values, or key-value pairs.
- Dictionaries maintain the insertion order of items.

```python
# Generate a dictionary with first name, last name, age, heart rate
print("1. Generate a dictionary with first name, last name, age, heart rate")
person = {'first_name': 'Anna', 'last_name': 'Rogers', 'age': 33, 'heart_rate': 65}
print(person)
print(" ")

# Return the keys of the dictionary
print("2. Return the keys of the dictionary")
print(person.keys())
print(" ")

# Return the values of the dictionary
print("3. Return the values of the dictionary")
print(person.values())
print(" ")

# Return the value for the "First Name" key
print("4. Return the value for the 'First Name' key")
print(person['first_name'])
print(" ")

# Use the get method to return the value for the "Last Name" key
print("5. Use the get method to return the value for the 'Last Name' key")
print(person.get('last_name'))
print(" ")

# Generate a dictionary using the dict function
print("6. Generate a dictionary using the dict function")
my_dict = dict([(1, 'A'), (2, 'B'), (3,'C'), (4,'D')])
print(my_dict)
print(" ")

# Return a list of the values of the dictionary
print("7. Return a list of the values of the dictionary")
print(list(my_dict.values()))
print(" ")
```

# <span style="color: teal;">Working with Sets</span>
### <span style="color: pink;">Characteristics of Sets</span> 
- They do not maintain any order of elements.
- They are mutable.
- They cannot contain duplicate items.
- They cannot be indexed or sliced.
- Elements must be immutable.

```python
# Generate set with the elements 1, 2, 3, 4,4,4,4,1
print("1. Generate set with the elements 1, 2, 3, 4,4,4,4,1")
set_1 = {1, 2, 3, 4, 4, 4, 4, 1}
print(set_1)
print(" ")

# Use the set function instead
print("2. Use the set function instead")
set_1 = set([1, 2, 3, 4, 4, 4, 4, 1])
print(set_1)
print(" ")

3 # Add the element 7 to the set
print("3. Add the element 7 to the set")
set_1.add(7)
print(set_1)
print(" ")

# Remove the element 4 from the set
print("4. Remove the element 4 from the set")
set_1.remove(4)
print(set_1)
print(" ")

# First set
print("5. First set")
set_A = {'a', 'a', 'a', 'b', 'c', 'd', 'e', 'f', 'e', 'a'}
print(set_A)
print(" ")

# Second set
print("6. Second set")
set_B = {'a', 'c', 'c', 'f', 'g'}
print(set_B)
print(" ")

# Return the union of the two sets using the union method
print("7. Return the union of the two sets using the union method")
print(set_A.union(set_B))
print(" ")


# Return the intersection of the two sets using the intersection method
print("8. Return the intersection of the two sets using the intersection method")
print(set_A.intersection(set_B))
print(" ")

# Return the union of the two sets using the pipe operator
print("9. Return the union of the two sets using the pipe operator")
print(set_A | set_B)
print(" ")

# Return the intersection of the two sets using the & operator
print("10. Return the intersection of the two sets using the & operator")
print(set_A & set_B)
print(" ")

# Return the difference between the two sets using the difference method
print("11. Return the difference between the two sets using the difference method")
print(set_A.difference(set_B))
print(" ")

# Reverse the order of the set difference
print("12. Reverse the order of the set difference")
print(set_B.difference(set_A))
print(" ")

# Return the symmetric difference between the two sets using the symmetric_difference method
print("13. Return the symmetric difference between the two sets using the symmetric_difference method")
print(set_A.symmetric_difference(set_B))
print(" ")
```
