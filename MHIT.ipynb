# Filtering data
- finding values with multiple characteristics
- use loc method try this with data set

```python
# find values of people with a current or past mental health disorder at companies that are primarily tech
tech_mental_health = osmi.loc[
    (osmi['Is your employer primarily a tech company/organization'] == True) & 
    (osmi['Have you had a mental health disorder in the past'] == "Yes")]
tech_mental_health.head()
```

```python
# break down into tech companies with mental health disorders
tech_mental_health = osmi.loc[
    (osmi['Is your employer primarily a tech company/organization'] == True) & 
    (osmi['Have you had a mental health disorder in the past'] == "Yes")]
tech_mental_health.head()
```

```python
# break down into tech with no mental health disorders
tech_no_mental_health = osmi.loc[
    (osmi['Is your employer primarily a tech company/organization'] == True) & 
    (osmi['Have you had a mental health disorder in the past'] == "No")]
tech_no_mental_health.head()
```

```python
# break down into non-tech companies with mental health disorders
non_tech_mental_health = osmi.loc[
    (osmi['Is your employer primarily a tech company/organization'] == False) & 
    (osmi['Have you had a mental health disorder in the past'] == "Yes")]
non_tech_mental_health.head()
```

```python
# break down into non-tech companies with no mental health disorders
non_tech_no_mental_health = osmi.loc[
    (osmi['Is your employer primarily a tech company/organization'] == False) & 
    (osmi['Have you had a mental health disorder in the past'] == "No")]
non_tech_no_mental_health.head()
```

```python
# display counts of each category
tech_mental_health_count = len(tech_mental_health)
tech_no_mental_health_count = len(tech_no_mental_health)
non_tech_mental_health_count = len(non_tech_mental_health)
non_tech_no_mental_health_count = len(non_tech_no_mental_health)
# make a neat table with these counts
data = {'Tech Company': [tech_mental_health_count, tech_no_mental_health_count], 'Non-Tech Company': [non_tech_mental_health_count, non_tech_no_mental_health_count]}
index = ['Mental Health Disorder', 'No Mental Health Disorder']
df = pd.DataFrame(data, index=index)
df
```
